```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
mins = np.array([2.31, 1.95, 1.88, 2.43, 2.7, 2.81])
maxes = np.array([3.96, 4.04, 4.24, 3.96, 4.14, 4.41])
means = np.array([3.306, 3.345, 3.393, 3.338, 3.353, 3.415])
std = np.array([0.298, 0.306, 0.309, 0.295, 0.260, 0.233])

# create stacked errorbars:
plt.errorbar(np.arange(6), means, std, fmt='ok', lw=3)
plt.errorbar(np.arange(6), means, [means - mins, maxes - means], ecolor='gray', lw=1)
plt.xlim(-1, 6)
plt.xlabel('Cluster number')
plt.ylabel('Untappd rating')
plt.ylim(1.5, 4.5)
plt.title('Errorbar for K-means clustering based on the colour histogram')
```
